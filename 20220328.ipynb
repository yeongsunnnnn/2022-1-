```python
def function(addr):
  URL = ('http://apis.data.go.kr/B552584/MsrstnInfoInqireSvc/'
      'getMsrstnList?'
      'addr=서울'
      '&pageNo=1'
      '&numOfRows=100'
      '&serviceKey=7P3DUyi4Jfok7JvvfXOURv5s6FK12EfWBU45yI2KPrDGrzqYFwUwHDowgzMC64xi2kbKE%2FSsWRpKEvpJdctp4A%3D%3D'
      '&returnType=json')
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  addr = [item['addr'] for item in items]
  dmX = [item['dmX'] for item in items]
  dmY = [item['dmY'] for item in items]
  mangName = [item['mangName'] for item in items]
  stationName = [item['stationName'] for item in items]
  year = [item['year'] for item in items]

  addr = pd.Series(addr)
  dmX = pd.Series(dmX)   
  dmY = pd.Series(dmY)
  mangName = pd.Series(mangName)
  stationName = pd.Series(stationName)
  year = pd.Series(year)

  df = pd.DataFrame({'addr':addr, 'dmX':dmX, 'dmY':dmY, 'mangName':mangName, 'stationName':stationName, 'year':year})
  return df
```

```python
iface = gr.Interface(function, gr.inputs.Dropdown(['동작구', '성동구', '중랑구', '강동구', '관악구', '구로구', '중구', '양천구', '금천구', '서대문구', '관악구', '성북구', '노원구', '동대문구', '광진구', '강남구', '강서구']), outputs=['dataframe'])
iface.launch(debug=True)
```
